```python
player_stats = pd.read_csv("2022-2023_NBA_Player_Stats_-_Regular.csv", delimiter=";", encoding="latin-1", index_col=0)

df = player_stats

df.head()
```

```python
TS = ((df.PTS) / (2 * ((df.FGA) + (0.44 * df.FTA)))) * 100

df.insert(29, "TS%", TS)

df.head()
```

```python
center = df[df.Pos == "C"]
point_guard = df[df.Pos == "PG"]
shooting_guard = df[df.Pos == "SG"]
small_forward = df[df.Pos == "SF"]
power_forward = df[df.Pos == "PF"]

small_forward.head()
```

```python
PO_dupl = df[df.duplicated(subset='Player', keep=False)]
print("Duplicate rows:")
print(PO_dupl)
print("")

num_duplicates = len(PO_dupl['Player'].unique())
print("The number of players recorded two or more times in the dataset is: " + str(num_duplicates))

df.drop_duplicates(subset='Player', inplace=True)
print("DataFrame after dropping duplicates:")
print(df)
```

```python
print("Above average scoring centers: ")
good_center = df.Player[(df["TS%"] >= 60.99) & (df.G > 54) & (df.Pos == "C") & (df.PTS >= 12.1)]

print("Above average scoring SFs: ")
good_sf = df.Player[(df["TS%"] >= 60.99) & (df.G > 54) & (df.Pos == "SF") & (df.PTS >= 12.1)]

print("Above average scoring SGs: ")
good_sg = df.Player[(df["TS%"] >= 60.99) & (df.G > 54) & (df.Pos == "SG") & (df.PTS >= 12.1)]

print("Above average scoring PGs: ")
good_pg = df.Player[(df["TS%"] >= 60.99) & (df.G > 54) & (df.Pos == "PG") & (df.PTS >= 12.1)]

print("Above average scoring PFs: ")
good_pf = df.Player[(df["TS%"] >= 60.99) & (df.G > 54) & (df.Pos == "PF") & (df.PTS >= 12.1)]

```

```python
future = pd.concat([good_center, good_sf, good_sg, good_pg, good_pf], axis=1, join="outer")
future = future.dropna(how='all', axis=0).dropna(how='all', axis=1)
future.index = np.arange(1, len(future) + 1)

future
```
